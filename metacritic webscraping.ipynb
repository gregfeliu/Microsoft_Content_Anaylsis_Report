```python
#create a function to pull results from one page

def scrape_metacritic_page(page_number, headers):
    num = 0
    try:
        page = requests.get(url='https://www.metacritic.com/search/movie/results?genres%5Bsci-fi%5D=1&search_type=advanced&page={}'.format(page_number), headers=headers)
        if page.status_code != 200:
            print(page.status_code)
        soup = BeautifulSoup(page.content, 'html.parser')
        results_page = soup.find_all(class_='main_stats')
        return results_page
    except requests.ConnectionError as e:
        print("OOPS!! Connection Error. Make sure you are connected to Internet. Technical Details given below.\n")
        print(str(e))
        print(page_number)
        pass
    except requests.Timeout as e:
        print("OOPS!! Timeout Error")
        print(str(e))
    except requests.RequestException as e:
        print("OOPS!! General Error")
        print(str(e))
    except KeyboardInterrupt:
        print("Someone closed the program") 
```

```python
results = scrape_metacritic_page(page_number, headers)
```

```python
#create a for loop to add movie title and score to a list of tuple - one page of results
def parse_results(results_page):
    master_movie_list = []
    for tag in results_page:
        new_tag = tag.text
        score = new_tag[:3].replace('\n','')
        title_tag = new_tag[3:-11].strip()
        title = title_tag[:-11].strip().replace('\n','')
        movie_tuple = (title,score)

        master_movie_list.append(movie_tuple)
    return master_movie_list

    
    

    
   
```
